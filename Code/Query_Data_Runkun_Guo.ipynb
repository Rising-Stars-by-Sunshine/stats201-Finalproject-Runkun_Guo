```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/stats201-PS2-Runkun_Guo/main/data/E0.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/stats201-PS2-Runkun_Guo/main/data/E1.csv')
df2
```

```python
def fun1(df):
    df["Total Score"] = df["FTHG"] + df["FTAG"]
    df["Total Shooting"] = df["HS"] + df["AS"]
fun1(df)
fun1(df2)
```

```python
df1=df[["Date","Time","Total Score","Total Shooting"]]
df3=df2[["Date","Time","Total Score","Total Shooting"]]
data = pd.merge(df1, df3, how ='outer', on =["Date","Time","Total Score","Total Shooting"])
data = data.dropna()
data
```

```python
data["MatchTime"] = data["Date"].str.cat(data["Time"], sep=" ")
data 
```

```python
# Convert object type to datetime object
data['MatchTime'] = pd.to_datetime(data['MatchTime'])
data = data.sort_values(by='MatchTime') 
print(data)
```
