```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create dataframe from data
# data = pd.read_csv("/Users/james/Desktop/Final Project/ZoomData.csv")
data = pd.read_csv("ZoomData.csv") # change it
df = pd.DataFrame(data)
# convert date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date']
```

```python
plt.plot(data["Date"], data["Volume"])
plt.show()     
```

```python
# split data before and after June 1st, 2022
before_df = df[df['Date'] < '2020-03-03']
after_df = df[df['Date'] >= '2020-03-03']
```

```python

# Draw the broken line statistical graph
plt.plot(df['Date'], df['Volume'], label='Daily Volume')

# Draw the fitting line before 2022.6.1
before_fit = np.polyfit(before_df.index, before_df['Volume'], 1)
before_line = np.poly1d(before_fit)
plt.plot(before_df['Date'], before_line(before_df.index), label='Before')

# Draw the fitting line after 2022.6.1
after_fit = np.polyfit(after_df.index, after_df['Volume'], 1)
after_line = np.poly1d(after_fit)
plt.plot(after_df['Date'], after_line(after_df.index), label='After')

plt.axvline(x=pd.to_datetime('2020-03-03'), color='red', linestyle='-', label='03/03/2020 Security Event')

# Add legend and labels
plt.legend()
plt.xlabel('Date')
plt.ylabel('Daily Volume')
plt.title('Daily Volume by Date')

# Show the plot
plt.show()
```
