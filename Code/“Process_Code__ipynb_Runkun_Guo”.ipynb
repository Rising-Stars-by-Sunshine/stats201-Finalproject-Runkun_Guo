# Part I: Import and Inspect Data


```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
google = pd.read_csv('/Users/james/Desktop/ZoomData.csv')
google.head()
```

# Part II: Prepare the Y varible for Regression

## 2.1. Write functions to calculte the Y variable for Regression

```python
#google['Open_past1'] =google['Open'].shift(1)
google['ROI']= (google['Close/Price']-google['Open'])/google['Open']
google
```

## 2.2. Make Sure that the Data Type of Y is "numeric"

```python
google['ROI']=pd.to_numeric(google['ROI'])
google.dtypes
```

# Part III: Prepare the Y variable for Classification

```python
cut=0
```

```python
google['Positive']=google['ROI']>=0
google.loc[(google['ROI'] >= cut), 'Positive'] = 1
google.loc[(google['ROI'] <cut), 'Positive'] = 0
google.head(20)
```

#Part IV: Create the X variables

## 4.1. Shift the Y to get past values

```python
google['ROI_past']=google['ROI'].shift(1)
google.head()
```

## 4.2. Calculate the Moving Averages

### Define the window

```python
window=100
```

```python
google['ROI_past_ma10']=google['ROI_past'].rolling(window=window,min_periods=10).mean()
google.head(20)
```

# Part V Train and Test Split

```python
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit()
print(tss)
```

```python
# change the train and test split parameters 
tss = TimeSeriesSplit(gap=0, max_train_size=None, n_splits=2, test_size=None)
```

```python
for train_idx, test_idx in tss.split(google):
    print("TRAIN:", train_idx, "TEST:", test_idx)
```

```python
train_idx
```

```python
test_idx
```

```python
train_google = google.filter(items=train_idx, axis=0)
test_google =  google.filter(items=test_idx, axis=0)
```
