```python
import pandas as pd
df = pd.read_csv('/Users/james/Desktop/PS2/E0.csv')
df2 = pd.read_csv('/Users/james/Desktop/PS2/E1.csv')
df2
```

```python
def fun1(df):
    df["Total Score"] = df["FTHG"] + df["FTAG"]
    df["Total Shooting"] = df["HS"] + df["AS"]
fun1(df)
fun1(df2)
```

```python
df2[["Date","Time","Total Score","Total Shooting"]]
```

```python
df[["Date","Time","Total Score","Total Shooting"]]
```

```python
df1=df[["Date","Time","Total Score","Total Shooting"]]
df3=df2[["Date","Time","Total Score","Total Shooting"]]
data = pd.merge(df1, df3, how ='outer', on =["Date","Time","Total Score","Total Shooting"])
data = data.dropna()
data
```

```python
# time_test = df['Date'].astype(str) + " " + df['Time'].astype(str)  # 日期加时间拼接
# data["MatchTime"] = pd.DataFrame(time_test)# 转化为 时间格式
data["MatchTime"] = data["Date"].str.cat(data["Time"], sep=" ")
data 
```

```python
# Convert object type to datetime object
data['MatchTime'] = pd.to_datetime(data['MatchTime'])
data = data.sort_values(by='MatchTime') 
print(data)
```

## Train and Test Split

```python
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit()
print(tss)
     
```

```python
# change the train and test split parameters 
tss = TimeSeriesSplit(gap=0, max_train_size=None, n_splits=2, test_size=None)
```

```python
for train_idx, test_idx in tss.split(df):
    print("TRAIN:", train_idx, "TEST:", test_idx)
```

```python
train_idx
```

```python
train_df = data.filter(items=train_idx, axis=0)
test_df =  data.filter(items=test_idx, axis=0)
```
